```python
import xgboost as xgb
import pandas as pd
import numpy as np

def predict(route, bkOrg, flDur, exBagg, lenStay, xgbModel):
    # Load the xgboost model
    bst = xgb.Booster()
    bst.load_model(xgbModel)

    features = ['route', 'booking_origin', 'flight_duration', 'wants_extra_baggage', 'length_of_stay']
    
    # Prepare the input data    
    data = [[route, bkOrg, flDur, exBagg, lenStay]]
    
    for col in data.select_dtypes("String"):
        data[0,col],_ = data[0,col].factorize()
        
    data = pd.get_dummies(data, columns = features)
    
    dmatrix = xgb.DMatrix(data)

    # Make predictions
    preds = bst.predict(dmatrix)

    return preds.tolist()[0]
```

# Convert model from json to tensorflow lite model

```python
import tensorflow as tf
import json

# Load the JSON model
with open('cust_book_pred.json', 'r') as json_file:
    json_model = json.load(json_file)

# Reconstruct the model using TensorFlow
model = tf.keras.Sequential()

# Example: Reconstruct layers from JSON
for layer_info in json_model['learner']['attributes']['model']['trees'][0]:
    if layer_info['type'] == 'Dense':
        model.add(tf.keras.layers.Dense(layer_info['units'], activation=layer_info['activation']))

# Compile the model (optional)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Convert the model to TensorFlow Lite format
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

# Save the TensorFlow Lite model
with open('cust_book_pred.tflite', 'wb') as tflite_file:
    tflite_file.write(tflite_model)
```
